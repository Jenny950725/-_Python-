### 打招呼 有名字和地區

```python
name = "佳宸"
place = "桃園"
```

```python
print("你好,來自" + place + "的" + name)
```

```python
print("你好,來自{}的{}".format(place,name))
```


1.不指定位置，照原本順序放到字串裡

2.設定指定位置，可不照順序，編號由0開始

```python
name = "佳宸"
place = "桃園"
age = "15"
#1.
message1 = "你好,來自{}的{}歲的{}!".format(place,age,name)
print(message1)
```

```python
#2-1. 記得由0開始數
message2 = "你好,來自{0}的{1}歲的{2}!".format(place,age,name)
print(message2)
```

```python
#2-2. 不照順序
message3 = "你好,來自{2}的{1}歲的{0}!".format(name,age,place)
print(message3)
```

## 匯率換算 
用format不用擔心資料型態

資料型態轉換

```python
m = 25
print(str(m) + "合台幣" + str(m * 30.5) + "元")
```

用format就不用

```python
m = 25
print("{}合台幣{}元".format(m,m*30.5))
```
