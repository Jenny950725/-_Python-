```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,100)
y = 1.2*x + 0.8 + 0.5 * np.random.randn(100)
plt.scatter(x,y)
```

```python
from sklearn.model_selection import train_test_split
```

#### 把原來的x,y分成80%的training data和20%的testing data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,
                                                test_size=0.2,
                                                random_state=40)
```

###### 分割training data 和 testing data的方法

訓練資料(輸入),測試資料(輸入),訓練資料(輸出),測試資料(輸出)=train_test_split(x,y,test_size=給測試資料的比例,random_state=隨便輸一個數字)

random_state : 讓資料更亂

線性回歸的機器

```python
from sklearn.linear_model import LinearRegression
```

```python
regr = LinearRegression()  #開機器
```

```python
regr.fit(x_train,y_train)
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,regr.predict(x_train),"r")
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,regr.predict(x_test),"r")
```

```python
plt.plot(x_train,regr.predict(x_train),"r")
plt.plot(x_test,regr.predict(x_test),"g")
```
