```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

numpy.random.rand(d0, d1, …, dn) 的隨機樣本位於[0, 1)中。

```python
x = np.random.rand(100,2)
```

```python
x[0]
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
clf = KMeans(n_clusters=3)
```

```python
clf.fit(x)
```

```python
clf.labels_
```

```python
plt.scatter(x[:,0],x[:,1],c=clf.labels_)
```

```python
X, Y = np.meshgrid(np.arange(0,1,0.02),np.arange(0,1,0.02))
```

```python
plt.scatter(X,Y)
```

```python
X
```

```python
predicted = clf.predict(np.c_[X.ravel(),Y.ravel()])
```

```python
z = predicted.reshape(X.shape)
```

```python
plt.scatter(X,Y,c=z,cmap=plt.cm.rainbow)
```
