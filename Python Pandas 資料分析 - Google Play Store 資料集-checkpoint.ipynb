## 資料清理

```python
clear_condition = df["Rating"]<=5
```

```python
data_cleared = df[clear_condition]
```

```python
data_cleared.head()
```

```python
data_cleared.iloc[10472]  #已經被清掉了
```

```python
print("平均數: ",data_cleared["Rating"].mean())
print("中位數: ",data_cleared["Rating"].median())
print("評分前100高的平均數: ",data_cleared["Rating"].nlargest(100).mean())
print("評分前1000高的平均數: ",data_cleared["Rating"].nlargest(1000).mean())
```

## 分析資料

安裝次數Installs

```python
df["Installs"]
```

object
在pandas裡面是字串的型態

pd.to_numeric()可以把資料型態轉成數字

```python
pd.to_numeric(df["Installs"])
```

出現錯誤，因為 , + 這些符號沒辦法被轉成數字型態

要用replace把這些東西換掉

```python
pd.to_numeric(df["Installs"].str.replace("[,+]", "")) 
```

```python
df.iloc[10472] #Installs 還真的是free ，要把它清掉
```

```python
df["Installs"] = pd.to_numeric(df["Installs"].str.replace("[,+]", "").replace("Free", "")) 
```

```python
df["Installs"]
```

算平均數

```python
df["Installs"].mean()
```

依據使用者輸入的關鍵字，顯示包含此關鍵字的APP名稱及數量

忽略大小寫: case = False 

```python
keyword = input("請輸入欲查找的關鍵字: ")
condition = df["App"].str.contains(keyword,case = False)
print("以下為符合關鍵字的搜索結果: \n",df["App"][condition])
print("共有",df["App"][condition].shape[0],"筆")
```
