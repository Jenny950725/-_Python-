```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
diabetes.feature_names
```

```python
print(diabetes.DESCR)
```

```python
只看BMI和
```

```python
X = diabetes.data
Y = diabetes.target
```

```python
X
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.5, random_state=55)
```

```python
regr = LinearRegression()
```

```python
regr.fit(x_train, y_train)
```

```python
y_predicted = regr.predict(x_test)
```

```python
y_predicted
```

```python
plt.scatter(y_test,y_predicted,s=30)
plt.plot([0,340],[0,340],"r")
plt.xlabel("y_test")
plt.ylabel("y_predicted")
```

```python
畫出 各特徵 和 一年後患糖尿病的指標數 之間的關係
```

```python
plt.figure(figsize = (12,12)) #圖的大小
for i,feature in enumerate(diabetes.feature_names):
    plt.subplot(5,3,i+1)
    plt.scatter(X[:,i],Y,s=1)
    plt.xlabel(feature)
    plt.ylabel('diabetes') 
    plt.tight_layout()  #自動調整圖的位置
```
