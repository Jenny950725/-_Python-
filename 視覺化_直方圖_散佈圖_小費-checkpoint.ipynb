## 直方圖: distplot()

kde是核密度函數，也就是把直方圖轉成折線圖，更方便讀者理解
https://datama.com.tw/20191112b1/

```python
sns.distplot(tips["total_bill"],kde=False)
plt.xlabel("total_bill")
plt.ylabel("number of people")
```

```python
sns.distplot(tips["total_bill"],kde=True)
#可看出消費額集中在接近20塊
```


打distplot出現的警告，原來2020年的書已經老惹...

### bins
直方圖是由一個個的區間來呈現的，我們可以藉由調整bins，來讓圖呈現更精細或更廣略的結果

```python
sns.distplot(tips["total_bill"],kde=False,bins=3) #全部只分成三塊 廣略
plt.xlabel("total_bill")
plt.ylabel("number of people")
```

```python
sns.distplot(tips["total_bill"],kde=False,bins=15) #全部分成十五塊 精細
plt.xlabel("total_bill")
plt.ylabel("number of people")
```

### vertical
讓圖轉90度，直方圖變橫條圖

用displot寫vertical不行

用distplot才行

```python
sns.distplot(tips["total_bill"],kde=False,bins=15,vertical=True)
plt.xlabel("number of people")
plt.ylabel("total_bill")
```

## 把兩張直方圖疊在一起: 午餐晚餐的普遍花費比較
### 篩選資料

```python
#兩種篩資料的方式
choice1 = tips["time"]=="Lunch"
lunch_tips = tips[choice1]
dinner_tips = tips[tips["time"]=="Dinner"]
```

```python
sns.distplot(lunch_tips["total_bill"],kde=False,color="r")
sns.distplot(dinner_tips["total_bill"],kde=False,color="g")
plt.xlabel("total_bill")
plt.ylabel("number of people")
plt.legend(labels=["lunch","dinner"])
```

質化變數
## Countplot()

```python
sns.countplot(x="smoker",data=tips)
plt.xlabel("smoker")
plt.ylabel("number of people")
```

```python
sns.countplot(x="sex",data=tips)
plt.xlabel("gender")
plt.ylabel("number of people")
```

## hue參數 
分類顯示

```python
sns.countplot(x="smoker",data=tips,hue="sex")
plt.xlabel("smoker")
plt.ylabel("number of people")
```

```python
sns.countplot(data=tips,x="day")
plt.xlabel("day")
plt.ylabel("number of people")
```

```python
sns.countplot(x="day",hue="sex",data=tips)
plt.xlabel("day")
plt.ylabel("number of people")
sns.despine(top=True,right=True)
```

# 散佈圖

```python
sns.scatterplot(x=tips["total_bill"], y=tips["tip"], data="tips")
```

hue指定欄位名稱

alpha 可設定0~1，控制不透明度，避免太多點重疊看不清楚

```python
sns.scatterplot(x=tips["total_bill"], y=tips["tip"], hue=tips["time"], alpha=0.7, data="tips")
```

style可設定不同欄位用不同形狀表示

```python
sns.scatterplot(x=tips["total_bill"], y=tips["tip"], hue=tips["time"], style=tips["time"], data="tips")
```
