```python
%matplotlib inline
import wordcloud
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from janome.tokenizer import Tokenizer
import pandas as pd
import seaborn as sns
```

```python
text = """Self-intro
I am a Taiwanese senior high school girl from Taoyuan. I like learning English and challenging myself by taking official 
English exams although I have only taken GEPT elementary test in my junior high and intermediate test just a few weeks ago.
After finishing the intermediate primary examination, it is likely that I am going  to face the re-examination. 
(aka "Speaking and Writing test")
That's why I'm here to practice and improve my writing.

Purpose of writing
In order to improve my English writing, I have decided to write a blog on the platform "Medium" after searching online. 
You probably want to ask a question "So, your article basically won't be seen by anybody online because of the algorithm, 
why do you still want to upload articles online?". I have to state that even though it's hard for my friends in reality 
to know that I am a shy and introverted girl, being afraid of sharing my works with friends. Hence, the Internet is a 
good place for me to practice without worrying about whether anybody is judging me and still has enough push to force 
me to practice my writing. The algorithm also helps. I am quite sure that no one can find my work unless I show them 
on my own accord. Though I am not prepared yet, I have the confidence that one day when the time is ripe , I can show
my writing to friends and also feel comfortable in my own skin.

Why medium?
I often search online when learning to code, and the articles with high quality are always the most numerous source 
of knowledge for me.That's how I discovered this platform and became attracted to it.
There are three main reasons for me to decide to upload my articles here.
	1. Simple network interface
	2. No advertisement
	3. None of my friends can find my works here

Plan
Talking about my writing plan, I'm thinking about updating weekly. On the one hand, senior high school life is really 
a struggle for me. I have to study really hard on my school work to achieve the standard I've set for myself, practice
the Piano and the Euphonium(well I have to admit that I RARELY spend time on music) and learn to code in my free time.
Sparing my limited time to write articles might be a tough task. I would rather prolong the preparation time for myself
than brave the risk of exhausting myself.On the other hand, I want the articles to be in smooth English, correct grammar
and kind of interesting so it must take time to draft and revise. That's why I set the update pace for myself. Wish me a
good start!

Themes
Now comes to the part  to discuss my writing themes. In my articles, I will mainly talk about student's life and extend 
to my hobbies, such as coding and music. I can tell you a little secret. I kind of want to write a blog sharing coding 
ideas for a long time. I want to write from a beginner's perspective. Coding isn't hard as it seems to be, while it is 
still not easy at the commencement. I always wonder if there is one guy who completely understands my struggles when I 
am learning to code. Maybe my path of learning would be easier and happier if the guy really exists. Therefore, even if my 
writing is neither professional nor skillful, I will try sharing code knowledges and my struggles on the path of learning.

I hope I can continually practice English writing whether I pass the elementary exam or not.
Wish me good luck!

"""
plt.figure(figsize=(16,8))
wordcloud_article = wordcloud.WordCloud(width=2000, height=1200, background_color="white")
wordcloud_article.generate(text) #
plt.imshow(wordcloud_article) #
plt.axis("off") #x,y
```
