字典名字 = {鍵1:值1,鍵2:值2}

```python
mydic = {'興趣':'學程式','愛吃的食物':'甜甜圈'}  
```

```python
mydic
```

查字典  字典名字[物件]

```python
mydic['興趣']
```

不能拿值去查鍵

```python
mydic['甜甜圈']   
```

加字

```python
mydic['愛好'] = '看金家雜貨店'
```

```python
mydic
```

把物件一個一個印出來

```python
for i in mydic:
    print("{}:{}".format(i,mydic[i]))
```

來查一個不存在的條目

```python
mydic['職業']
```

用method，沒有這個鍵的話，會顯示none

```python
print(mydic.get('職業')) 
```

可以自己改成任意的字，沒有找到這個鍵時會出現

```python
print(mydic.get('職業','錯誤訊息'))  
```

可以把一整個字典加到原本的字典中

```python
classmate = {'Joan':'1','Jenny':'2','Kate':'3'}
```

轉學生

```python
newClassmate = {'Jason':'4','Emily':'5','Bob':'6'}
```

```python
classmate.update(newClassmate)
```

```python
print(classmate) #把第二個字典接在第二個字典後面
```

刪除資料

```python
del classmate['Joan']
```

```python
print(classmate)
```

pop可以回傳被刪掉的值

```python
classmate.pop('Jenny')  
```

```python
print(classmate)
```
