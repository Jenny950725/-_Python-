```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

先建一個內容有正有負的array

```python
L = np.array([1,50,-45,-34,44,-66])
```

要找出小於0的數字

只有說他是不是小於0

```python
L<0
```

要把它挑出來

array[條件]

```python
L[L<0] 
```

先畫一個y軸跨越正負的函數圖形

```python
x = np.linspace(-5,5,500)
y = np.sinc(x)
plt.plot(x,y)
```

把y大於0的部分另外標示出來

```python
plt.plot(x,y)
plt.plot(x[y>0],y[y>0],'3')
```
