# 用SVM來做分類

## 假設有四個點，被分為兩個類別
x是資料
y是分類的答案

```python
x = np.array([[1,3],[-5,2],[5,-4],[6,-3]])
y = np.array([1, 1, 2, 2])  #前兩個點是第一類，後兩個點是第二類
```

```python
plt.scatter([1,-5,5,6],[3,2,-4,-3],c=[1, 1, 2, 2])
```

```python
from sklearnex.svm import SVC
```

clf:classifier

開一台機器

```python
clf = SVC()
```

讓他學習

```python
clf.fit(x, y)
```

## 預測結果

predict()裡面放array或list

```python
clf.predict([1,3])  #這樣是不行的
```

```python
clf.predict([[1,3]]) #predict()裡面必須放一個array或者list的形式
```

```python
clf.predict(x)
```

```python
clf.predict([[11,0]])
```

```python
clf.predict([[22,3],[-4,7]])
```
