```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

讀入波士頓房價數據
scikit learn內建的真實世界數據 稱作"Toy Data"


https://www.twblogs.net/a/5b88433a2b71775d1cdb5398

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

用feature_names看到所有的變數名稱，這裡有13個變數(features)

```python
boston.feature_names
```

```python
X = boston.data
Y = boston.target  ## (正確答案)
```

```python
len(X)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=50)
```

```python
regr = LinearRegression()
```

```python
regr.fit(x_train,y_train)
```

```python
y_predict = regr.predict(x_test)
```

正確答案放x軸 預測答案放y軸


如果很準->點點會在對角線上

```python
plt.scatter(y_test,y_predict) 
plt.plot([0,50],[0,50],"r") #畫對角線
```

集中對角線附近

在對角線上:完全命中

從這張圖中看到: 
在高房價的部分會低估

```python
plt.plot(x,x**2)
plt.plot(x,x*(-3))
```

想畫兩張獨立的圖



subplot:https://steam.oxxostudio.tw/category/python/example/matplotlib-subplot.html#a1

```python
plt.subplot(321)
plt.plot(x,x*(-3))

plt.subplot(3,2,2)
plt.plot(x,x**2)

plt.subplot(3,2,3)
plt.plot(x,np.cos(x))

plt.subplot(3,2,4)
plt.plot(x,np.tan(x))

plt.subplot(3,2,5)
plt.plot(x,np.sin(x))

plt.subplot(3,2,6)
plt.plot(x,x+5)
```

```python
plt.figure(figsize = (16,8)) #圖的大小
for i,feature in enumerate(boston.feature_names):
    plt.subplot(3,5,i+1)
    plt.scatter(X[:,i],Y,s=1)
    plt.xlabel(feature)
    plt.ylabel('Price') 
    plt.tight_layout()  #自動調整圖的位置
```
