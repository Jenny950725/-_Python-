```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from ipywidgets import interact
```

np.linspace的用法，在1~5之間，均勻撒點

```python
n = np.linspace(1,5,5)
print(n)                  
```

```python
m = np.linspace(1, 5, 20)
print(m)
```

```python
x = np.linspace(0,2*np.pi,1000)
```

```python
def draw(n):
    y = np.sin(x*n)
    plt.plot(x,y,lw=1)
    plt.show()
```

```python
draw(4)
```

n是介於0到50之間的整數及小數

```python
interact(draw,n=(1.,50.))        
```

```python
h = np.linspace(0,100,1000)
def draw2(m):
    k = m*h
    plt.plot(h,k,lw=1)
    plt.show
```

```python
draw2(2)
```

```python
interact(draw2,m = (1,10))
```
