```python
df2 = pd.read_csv("student.csv")
```

```python
df2
```

新增總和欄位:dataframe名稱["新增的欄位名"] = dataframe名稱[要總和的欄位名稱list].sum()

1 代表 行的方向加總

0 代表 列的方向加總

```python
df2["總分數"] = df2[["國文","英文","數學"]].sum(1) 
```

```python
df2
```

刪除不要的欄位

```python
df2.drop(["總分數"],axis=1) 
```

新增一個經過計算的欄位: dataframe名稱["要新增的欄位名"] = dataframe名.項目 計算式

```python
df2["主科"] = df2.數學 * 2 + df2.英文
```

```python
df2
```

dataframe名.sort_values(by=根據甚麼項目排序,ascending=False) 

ascending=False 由大到小排序

ascending=True  或者不加 也可由小到大排序 

```python
df2.sort_values(by="總分數",ascending=False) 
```

dataframe名.sort_values(by=[根據那些項目排序],ascending=False) 

先用項目1排序 再用項目2 項目3... ...

```python
df2.sort_values(by=["總分數","主科"],ascending=False)
```
